```python
#Car Type Pie Chart
df.Type.value_counts().plot(kind = 'pie',autopct='%1.0f%%')
plt.ylabel('Car Type')
plt.axis('equal')
plt.show()
```

```python
#Price Category Pie Chart
df.Price_Category.value_counts().plot(kind = 'pie',autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#Retail Price vs Car Type
sns.swarmplot(df.Type,df.Retail_Price)
plt.xlabel('Car Type')
plt.ylabel('Retail Price')
plt.show()

```

```python
#Fuel Efficiency vs Car Type
sns.swarmplot(df.Type,df.City_MPG)
plt.xlabel('Car Type')
plt.ylabel('Mileage')
plt.show()
```

```python
#Fuel Efficiency by car type and price range
sns.barplot(data = df, x = 'Type', y='City_MPG', hue = 'Price_Category', ci = 2.0)
plt.xlabel("Car Type")
plt.ylabel("City Mileage")
plt.show()
```
